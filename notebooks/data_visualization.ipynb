```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize =(8,6))
# axes is a 2D array of Axes objects (e.g., axs[0,0], axs[0,1], etc.)

# line graph values 
x1=np.linspace(0,0.3,12)
y1= x1**2+(4*x1-2.1)

# Bar chart 
categories= ["a","b","c","d","e","f"]
values = [3,4,2,5,4,7]
plt.bar(categories,values)

#scatter plot values 

x2= [3,4,6,2,6,7]
y2= [34,43,56,76,65,34]

# Histogram value 
data= np.random.normal (0,1,100)

# Plot 1: Histogram (top-left)
axes[0, 0].hist(data, bins=5, edgecolor='black')
axes[0, 0].set_title('Histogram')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')

# Plot 2: Scatter Plot (top-right)
axes[0, 1].scatter(x2, y2, color='red')
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('X-axis')
axes[0, 1].set_ylabel('Y-axis')

# Plot 3: Bar Chart (bottom-left)
axes[1, 0].bar(categories, values, color='green')
axes[1, 0].set_title('Bar Chart')
axes[1, 0].set_xlabel('Category')
axes[1, 0].set_ylabel('Value')

# Plot 4: Line Graph (bottom-right)
axes[1, 1].plot(x1, y1)
axes[1, 1].set_title('Line Graph')
axes[1, 1].set_xlabel('X-axis')
axes[1, 1].set_ylabel('Y-axis')


# Adjust kayout and display all plots
plt.tight_layout()
plt.show()
```
