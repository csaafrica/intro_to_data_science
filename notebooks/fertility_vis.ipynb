# Fertility Data Visualization 


```python
#import required  libraries 
print("starting importing of libraries.")

import  numpy  as np  
import  pandas as pd
import  seaborn as  sns
import matplotlib.pyplot as plt

print("libraries imported successfully.")
```

```python
# load your data 

gapminder= pd.read_csv("../data/gapminder.csv")

# gapminder.describe()
# gapminder.head()
# bike_data.value_counts()
# gapminder.info()
gapminder.dtypes
```

## Introduction to Data Visualization 

```python
#The following cells introduce different plot types
gapminderYear = gapminder["year"].copy()
# gapminderYear.describe()
gapminderYear.unique()
```

```python
# extracting a potion of the data from the dataset. eg extract all the values from the year 2002 

# Create a mask that will help you filter out the data from 2002 only 
mask2002 = gapminder["year"] == 2002
mask2002
```

```python
# get the rows with data from 2002 only

get2002gapminder = gapminder[mask2002]

getKenyagapminder = gapminder["country"] == "Kenya"

# get2002gapminder
getKenyagapminder
```

```python
# Try extracting the values from only kenya 

maskKenya = (getKenyagapminder)&(mask2002) 

# get only the values of kenya in the year 2002 
# gapminder[(geKenyagapminder)&(mask2002)]
gapminder[maskKenya]
```


# Plotting Figures 

```python
# create a figure object 
gapminder.info()
fig, axes = plt.subplots(figsize =(8,6))
```

```python
# plot life expectancy as a gdp per capita [x-life expectancy y-gdp per capita]
fig, axes = plt.subplots(figsize =(10,6))

plt.scatter( gapminder["gdp_per_cap"],gapminder ["life_exp"], color='red')
# plt.scatter(x,y)
plt.ylabel("Life Expectancy")
plt.xlabel("GDP Per Capita")
# plt.ylabel("GDP Per Capita")
plt.title("Scatter Plot for showing Life Expectancy against GDP per Capita")

plt.show()
```

```python
# plot life expectancy in 2002 as a gdp per capita [x-life expectancy y-gdp per capita]
fig, axes = plt.subplots(figsize =(10,6))

plt.scatter( get2002gapminder["gdp_per_cap"],get2002gapminder["life_exp"], color='red')
# plt.scatter(x,y)
plt.ylabel("Life Expectancy")
plt.xlabel("GDP Per Capita")
# plt.ylabel("GDP Per Capita")
plt.title("Scatter Plot for showing Life Expectancy against GDP per Capita")

plt.show()
```

```python
# get the life expectancy against gdp per capita for Kenya 

# 1. get all the values from only Kenya by accessing the country value from gapminder 
Kenyagapminder = gapminder[gapminder["country"]== "Kenya"]

# Kenyagapminder

fig, axes = plt.subplots(figsize= (8,6))

plt.scatter( Kenyagapminder["gdp_per_cap"],Kenyagapminder["life_exp"], color='red')

plt.xlabel("GDP Per Capita")
plt.ylabel("Life Expectancy")

plt.title("Scatter Plot showing GDP per Capita against Life Expectancy in Kenya")

plt.show()
```
