### Once statisfied with the exploratory of data above we move to

```python
#performing a left join. picking from the left table and merging/updating with that on the right using primary key name(common item)
join_data = pd.merge(
    left=joining_data1,
    right=joining_data2,
    how = "left",
    on = "name"
)
join_data
```

```python
#performing right join
join_data_right = pd.merge(
    left=joining_data1,
    right=joining_data2,
    how = "right",
    on = "name"
)
join_data_right
```

```python
#performing inner join takes the common values that are shared by both joining_data1 and joining_data2 only
join_data_inner = pd.merge(
    left=joining_data1,
    right=joining_data2,
    how = "inner",
    on = "name"
)
join_data_inner
```

```python
#performing outer join-we are taking all the valuesfrom both the left and right
join_data_outer = pd.merge(
    left=joining_data1,
    right=joining_data2,
    how = "outer",
    on = "name"
)
join_data_outer
```
